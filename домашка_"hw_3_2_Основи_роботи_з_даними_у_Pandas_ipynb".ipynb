Як бачимо з попередньої задачі - відсоток інвойсів, тобто покупок в двох містах майже однаковий, але при цьому валовий дохід менший, тобто люди купують на меншу суму.

Можливі причини різниці доходу в різних містах:
1. Відсутність або менша кількість в асортименті товарів, які користуються більшим попитом
2. Більша кількість конкуренті/менше населення в місті з меншим доходом
3. Клієнтам не подобається сервіс в магазині.

Для перевірки першої гіпотези варто порівняти дані з колонок Product line та Quantity, тобто подивитись, що саме та в якій кількості купують клієнти та порівняти ці дані з інформацією про те, скільки саме кожного товару було в наявності в магазині. Тоді, очевидно, збільшити кількість товарів, які користуються найбільшим попитом.

Другу гіпотезу перевіряємо з відкритих джерел. Якщо магазинів-конкурентів дійсно більше - потрібно порівняти наскільки ми конкурентні порівняно з ними, можливо, вони мають привабливіші умови для покупок, різні акції, програми лояльності тощо і подумати, що ми можемо запровадити в себе. Перевірити Unit price, можливо, ціна на найпопулярніші товари зависока, більша, ніж у конкурентів.
Якщо в місті менше населення, тоді більш коректно буде порівнювати дохід пропорційно. А далі подумати, як залучити більшу кількість покупців - можливо, додати в асортимент якісь товари, які зараз купують в інших магазинах.

Третя гіпотеза перевіряється даними з колонки Rating. Якщо сервіс виявиться поганим - розбиратись детальніше. Можливо, в магазині працює лише одна каса, постійно великі черги і люди не хочуть витрачати свій час, тому йдуть туди, де менша черга. Або працює хамовитий персонал чи товари не встигають викладати. Тоді звісно, вирішувати конкретну причину: найняти більшу кількість працівників для того, аби працювало більше кас і товар викладався вчасно, провести навчання по роботі з покупцями тощо.
